```python
# Download der Tabelle als Flatfile-csv 
df73421_0002 = pd.read_csv("https://www-genesis.destatis.de/genesisWS/rest/2020/data/tablefile?username="+KENNUNG+"&password="+PASSWORT 
                      +"&name=73421-0002&startyear=1993&format=ffcsv", dtype=str, sep=';') 
                        # csv files per api are utf-8 encoded
df73421_0002
```

```python
bierabsatz = df73421_0002[["Zeit", "2_Auspraegung_Code", "ABS011__Absatz_von_Bier__hl"]]
```

```python
# Monate ohne Angabe abfiltern (typischerweise Rest des laufenden Jahres)            
bierabsatz = bierabsatz.loc[bierabsatz["ABS011__Absatz_von_Bier__hl"]!="..."]
```

```python
# Datums-Objekt im Format D/MM/YYYY erzeugen 
bierabsatz["date"] = "1/"+bierabsatz["2_Auspraegung_Code"].str[5:7] +"/"+ bierabsatz["Zeit"].astype(str)
```

```python
# Spaltenüberschriften für den csv-Output verkürzen
output = bierabsatz.drop(columns=["Zeit","2_Auspraegung_Code"]).rename(columns={"ABS011__Absatz_von_Bier__hl" : "bierabsatz"})
```
